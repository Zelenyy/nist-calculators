```python
from star import AlphaSTARCalculator, AlphaMaterials
```

```python
material = AlphaMaterials.BERYLLIUM
calculator = AlphaSTARCalculator(material)
```

```python
nuclear = calculator.calculate_nuclear_stopping_powers(10) 
electronic = calculator.calculate_electronic_stopping_powers(10)

total = calculator.calculate_total_stopping_powers(10)
print(total, nuclear+electronic)
```

```python
# Different energy arguments
calculator.calculate_csda_ranges() # use default
calculator.calculate_projected_range(10)
calculator.calculate_detour_factors([10,20,30])
```

```python
import tabulate
from IPython.core.display import HTML

data = calculator.calculate_table()

table = tabulate.tabulate(data, headers=(data.dtype.names), tablefmt="html")
HTML(table)
```
