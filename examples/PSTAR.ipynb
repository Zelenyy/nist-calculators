```python
material = ProtonMaterials.BERYLLIUM
calculator = ProtonSTARCalculator(material)
```

```python
nuclear = calculator.calculate_nuclear_stopping_powers(10) 
electronic = calculator.calculate_electronic_stopping_powers(10)

total = calculator.calculate_total_stopping_powers(10)
print(total, nuclear+electronic)
```
